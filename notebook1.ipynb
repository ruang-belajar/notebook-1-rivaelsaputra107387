```python
df = pd.read_csv("data-kebakaran.csv")
df.head()
```

**_Tugas 1:_**
Filter data laporan kebakaran yang sumber informasinya dari Masyarakat





```python
masyarakat_df = df[df['sumber_info'] == 'Masyarakat']
print("+ Text\nTugas 1:\n", masyarakat_df)
```

**_Tugas 2:_** Filter data laporan kebakaran yang sumber informasinya dari Masyarakat di bulan Februari

```python
februari_masyarakat_df = df[(df['sumber_info'] == 'Masyarakat') & (df['bulan'] == 2)]
print("+ Text\nTugas 2:\n", februari_masyarakat_df)
```

**_Tugas 3:_** Filter data laporan kebakaran dengan nilai kerugian lebih dari 100 juta

```python
kerugian_100juta_df = df[df['kerugian'] > 100000000]
print("+ Text\nTugas 3:\n", kerugian_100juta_df)
```

**_Tugas 4:_** Filter data laporan kebakaran selama semester 2 tahun 2022

```python
semester_2_2022_df = df[(df['bulan'] >= 7) & (df['bulan'] <= 12) & (df['bulan'] == 2022)]
print("+ Text\nTugas 4:\n", semester_2_2022_df)
```

**_Tugas 5:_** Filter data laporan kebakaran selama dari Maret sampai Juli 2022

```python
maret_juli_2022_df = df[(df['bulan'] >= 3) & (df['bulan'] <= 7) & (df['bulan'] == 2022)]
print("+ Text\nTugas 5:\n", maret_juli_2022_df)
```

**_Tugas 6 (duplikat dari Tugas 5):_** Filter data laporan kebakaran selama dari Maret sampai Juli 2022

**_Tugas 7:_** Filter data laporan kebakaran Rumah Tinggal yang disebabkan oleh Listrik

```python
rumah_listrik_df = df[(df['objek_awal'] == 'Rumah Tinggal') & (df['penyebab'] == 'Listrik')]
print("+ Text\nTugas 7:\n", rumah_listrik_df)
```
